```java
import java.util.List;
import java.util.ArrayList;
import java.util.Set;
import java.util.ArrayDeque;
import java.util.ArrayDeque;


public class CollectionsPractice {

    // Part 1: Working with Lists
    public static List<Integer> filterEvenNumbers(List<Integer> input) {
        List<Integer> result = new ArrayList<>();
        for (int i = 0; i < input.size(); i++) {
            int num = input.get(i);
            if (num % 2 == 0) {
                result.add(num);
            }
        }
        return result;
    }

    // Part 2: Working with Sets
    public static Set<String> findIntersection(Set<String> set1, Set<String> set2) {
        Set<String> result = new HashSet<>();
        for (String item : set1) {
            if (set2.contains(item)) {
                result.add(item);
            }
        }
        return result;
    }

    // Part 3: Working with Deques
    public static void simulateDequeOperations() {
        Deque<String> customerLine = new ArrayDeque<>();

        // Enqueue: Add 4 customers to the end
        customerLine.addLast("Please");
        customerLine.addLast("put");
        customerLine.addLast("the ");
        customerLine.addLast("fries");
        System.out.println("Initial line after adding 4 customers (end): " + customerLine);

        // VIP Arrival
        customerLine.addFirst("VIP 1st");
        System.out.println("After VIP arrival (added at front): " + customerLine);

        // Process Line
        customerLine.removeFirst();
        System.out.println("After serving the customer at the front (removed): " + customerLine);

        // Inspect Line
        System.out.println("Current front of the line: " + customerLine.peekFirst());
        System.out.println("Current back of the line: " + customerLine.peekLast());

        // Size Check
        System.out.println("Total number of customers waiting: " + customerLine.size());
    }

    public static void main(String[] args) {
        // Test Part 1: List
        List<Integer> inputList = new ArrayList<>();
        for (int i = 1; i <= 10; i++) {
            inputList.add(i);
        }
        List<Integer> evenNumbers = filterEvenNumbers(inputList);
        System.out.println("Input List: " + inputList);
        System.out.println("Filtered Even Numbers: " + evenNumbers);

        // Test Part 2: Set
        Set<String> set1 = new HashSet<>(Arrays.asList("apple", "banana", "cherry", "date"));
        Set<String> set2 = new HashSet<>(Arrays.asList("banana", "date", "fig", "grape"));
        Set<String> intersection = findIntersection(set1, set2);
        System.out.println("Set1: " + set1);
        System.out.println("Set2: " + set2);
        System.out.println("Intersection: " + intersection);

        // Test Part 3: Deque
        simulateDequeOperations();

        // Bonus Challenge Answer
        System.out.println("\nBonus Challenge: The best collection for storing unique student IDs with fast lookups is HashSet.");
    }
} 

CollectionsPractice.main(null)
```
