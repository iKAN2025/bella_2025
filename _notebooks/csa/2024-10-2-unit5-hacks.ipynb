# Hack 1

He then needs to organize the minions in terms of recovery time and usefulness. To do this, Gru needs you to make a minion class with the instance variables color, name, energy levels, gadgets, hair, height

Hack 5
Gru wants to make sure his workers are not overworked as per OSHA. So, Gru wants you to print out the average energy levels of all his Minions. (Hint: you should use static variables)

For 0.90+
Dr. Nefario is trying to assign a recovery time for each minion! Minions who were purple and got cured are very tired, and so are a lot of minions with low energy levels. Create a simple algorithm to calculate how long each minion needs to recover based on their color and energy levels.

```java
public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private String gadget;
    private String hair;
    private double height;

    // Default constructor
    public Minion() {
        this.color = "Yellow";      
        this.name = "Minion";       
        this.energyLevel = 100;     
        this.gadget = "Banana";     
        this.hair = "GREASY";        
        this.height = 1.2;         
    }

    // Parameterized constructor
    public Minion(String color, String name, int energyLevel, String gadget, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.gadget = gadget;
        this.hair = hair;
        this.height = height;
    }

    // Getters
    public String getColor() {
        return color;
    }

    public String getName() {
        return name;
    }

    public int getEnergyLevel() {
        return energyLevel;
    }

    public String getGadget() {
        return gadget;
    }

    public String getHair() {
        return hair;
    }

    public double getHeight() {
        return height;
    }

    // Setters
    public void setColor(String color) {
        this.color = color;
    }

    public void setName(String name) {
        this.name = name;
    }

    public void setEnergyLevel(int energyLevel) {
        this.energyLevel = energyLevel;
    }

    public void setGadget(String gadget) {
        this.gadget = gadget;
    }

    public void setHair(String hair) {
        this.hair = hair;
    }

    public void setHeight(double height) {
        this.height = height;
    }

    // Method to display minion information
    public void displayMinionInfo() {
        System.out.println("Minion Name: " + name);
        System.out.println("Color OF MINION: " + color);
        System.out.println("Energy Level: " + energyLevel);
        System.out.println("Gadget: " + gadget);
        System.out.println("Hair: " + hair);
        System.out.println("Height: " + height + " FEET");


        public static void main(String[] args) {
                // Using the default constructor
                Minion defaultMinion = new Minion();
                defaultMinion.displayMinionInfo();

                System.out.println(); // Just for better separation of output

                // Using the parameterized constructor
                Minion specificMinion = new Minion("Purple", "Bob", 80, "Laser", "Long", 1.5);
                specificMinion.displayMinionInfo();
            }
}



    }
}
Minion.main(null)
```
